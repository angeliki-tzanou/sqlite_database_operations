###Defining the numerical columns in the dataset to perform some data analysis:



> Replacing NaN values with the mean of the numerical columns to perform some basic statistical analysis and derive more accurate results:



```python
df_3.fillna(df_3.mean(), inplace=True)
```



>  Replacing NaN values with the mode of the categorical columns(if they have one) to perform some basic statistical analysis and derive more accurate results:




```python
df_3.columns
```

```python
categorical_col = df_3.select_dtypes(include=['category']).columns
```

```python
for col in categorical_col:
  mode = df_3[col].mode()[0]
  df_3[col].fillna(mode, inplace=True)
```



> Frequency counts for categorical columns:



- Creating a dictionary to store the frequency counts data throughout this loop command below

```python
frequency_counts = {}
cat_col = df_3.select_dtypes(include=['object','category']).columns

for col in cat_col:
  frequency_counts[col] = df_3[col].value_counts()
```

```python
for col, counts in frequency_counts.items():
  print(f"Frequency counts for '{col}':")
  print(counts)
  print()
```

From the above analysis it was very interesting to observe the frequency of certain CPT codes and procedures. Additionally the fact the outpatient encounters severely outnumbered the inpatient ones, was also very insightful, especially when various of the procedures would be considered highly complex.

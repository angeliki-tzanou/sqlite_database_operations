###Defining the numerical columns in the dataset to perform some data analysis:



>  Replacing NaN values with the mode of the categorical columns(if they have one) to perform some basic statistical analysis and derive more accurate results:




```python
df_3.columns
```

```python
categorical_col = df_3.select_dtypes(include=['category']).columns
```

```python
for col in categorical_col:
  mode = df_3[col].mode()[0]
  df_3[col].fillna(mode, inplace=True)
```



> Frequency counts for categorical columns:



- Creating a dictionary to store the frequency counts data throughout this loop command below

```python
frequency_counts = {}
cat_col = df_3.select_dtypes(include=['object','category']).columns

for col in cat_col:
  frequency_counts[col] = df_3[col].value_counts()
```

```python
for col, counts in frequency_counts.items():
  print(f"Frequency counts for '{col}':")
  print(counts)
  print()
```

From the above analysis it was very interesting to observe the frequency of certain CPT codes and procedures. Additionally the fact the outpatient encounters severely outnumbered the inpatient ones, was also very insightful, especially when various of the procedures would be considered highly complex.

##SQLite Database Operations:

```python
import sqlite3
import pandas as pd
```



> Creating the database:



```python
conn = sqlite3.connect('health.db')
c = conn.cursor()
```



> Manual Table Creation:



```python
c.execute('''CREATE TABLE Manual_Table2
                (
                    [hospital_name] text,
                    [insurance_type] text,
                    [ICD10_code] real,
                    [code_description] text,
                    [age] integer
                )''')

```

- Ensuring that the table has been created by printing the Name of the table:

```python
c.execute('''SELECT name FROM sqlite_master WHERE type='table' ORDER BY name;''')

for value in c.fetchall():
    print(value)
```

```python
## Now perform a Select * on the table to make sure it worked and it returns no values

c.execute('''SELECT * FROM Manual_Table2''')
print(c.fetchall())

conn.commit()
```

- Creating an INSERT INTO SQL query:

```python
c.execute('''
  INSERT INTO Manual_Table2 (hospital_name, insurance_type, ICD10_code, code_description, age)
  VALUES
    ('Mount Sinai Hospital', 'Empire BCBS', 'Z80.0', 'Family hx of malignant neoplasm of digestive organs', '33'),
    ('Lenox Hill Hospital', 'Emblem Health GHI', 'Z85.3', 'Personal hx of malignant neoplasm of the breast', '55')
''')
```

```python
c.execute('''SELECT * FROM Manual_Table2''')
print(c.fetchall())

conn.commit()
conn.close()
```
